## 1. Import NumPy

```
import numpy as np
```

## 2. Check NumPy Version

```
print(np.__version__)
```

## 3. Create a NumPy Array

```
arr = np.array([10, 20, 30, 40, 50])
print(arr)
print(type(arr))
```

## 4. Python List vs NumPy Array

```
python_list = [1, 2, 3, 4, 5]
numpy_array = np.array([1, 2, 3, 4, 5])

print(type(python_list))
print(type(numpy_array))
```

## 5. One-Dimensional Array

```
arr1 = np.array([1,2,3,4,5])
print(arr1)
```

## 6. Two-Dimensional Array

```
arr2 = np.array([[1,2,3],[4,5,6]])
print(arr2)
```

## 7. Three-Dimensional Array

```
arr3 = np.array([[[1,2],[3,4]],[[5,6],[7,8]]])
print(arr3)
```

## 8. Number of Dimensions

```
print(arr1.ndim)
print(arr2.ndim)
print(arr3.ndim)
```

## 9. Why NumPy?
- Faster than Python lists
- Less memory usage
- Supports multidimensional arrays
- Foundation for Pandas, SciPy and Machine Learning

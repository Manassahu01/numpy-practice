# 📦 Array Creation in NumPy

Learn how to create NumPy arrays of different dimensions and data types.

## 1. Import NumPy

```
import numpy as np
```

## 2. Create a 1D Array

```
arr1 = np.array([10, 20, 30, 40, 50])
arr1
```

## 3. Create a 2D Array

```
arr2 = np.array([[1,2,3],[4,5,6]])
arr2
```

## 4. Create a 3D Array

```
arr3 = np.array([[[1,2],[3,4]],[[5,6],[7,8]]])
arr3
```

## 5. Create Arrays with Different Data Types

```
int_arr = np.array([1,2,3], dtype=int)
float_arr = np.array([1,2,3], dtype=float)
str_arr = np.array(["Python","NumPy","Pandas"])

print(int_arr)
print(float_arr)
print(str_arr)
```

## 6. Create Array from Tuple

```
t = (10,20,30)
arr = np.array(t)
arr
```

## 7. Check Array Type

```
arr = np.array([1,2,3])
print(type(arr))
print(arr.dtype)
```
